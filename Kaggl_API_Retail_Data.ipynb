```python
#extract file from zip file
import zipfile

zip_ref = zipfile.ZipFile('orders.csv.zip') 
zip_ref.extractall() # extract file to dir
zip_ref.close() # close file
```

```python
import pandas as pd

df = pd.read_csv('orders.csv')
df
```

```python
# creating copy of dataframe for further analysis and restructuring 
df1 = df 
df1
```

```python
df1.info()

# order data is Object type convert into datetime
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'])
df1['Order Date']

df1.info()
```

```python
df1.columns = df1.columns.str.lower()
df1.columns
df1
#converting column name to lowercase
```

```python
df1.columns = df1.columns.str.replace(" ","_")
df1
# columns name are okay now
```

```python
df1.describe()
# no significance here as columns are categorical and datetime
```

```python
df1['segment'].unique()
# checking the viability of column
```

```python
df1['ship_mode'].unique()
# checking the viability of column
# 'Not Available', 'unknown', these value ar eof type NaN we have to make it as nan
```

```python
import numpy as np
df1['ship_mode'] = df1['ship_mode'].replace(['Not Available', 'unknown'], np.nan)

# using replace func to replace NA and unknown to null value
```

```python
df1['ship_mode'].unique()
```

```python
df1.head()
# now add additional columns for analysis
```

```python
df1['discount'] = df1['list_price']  * df1['discount_percent'] * 0.01
df1['selling_price'] = df1['list_price'] - df1['discount']
df1['profit'] = df1['selling_price'] - df1['cost_price'] 
df1[['discount','selling_price','profit']]
```

```python
df1.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
# doing inplace = true carry forward the changes to dataframe
```

```python
df1.to_csv('Kaggle_sales_data_ankit_bansal.csv')
```
